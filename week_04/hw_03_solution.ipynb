## Основной блок задач

```python
'''хороший тон импортировать все библиотеки в начале ноутбука, 
но тк как в нашем ноутбуке очень много тем, позволим себе исключение'''

from bs4 import BeautifulSoup
import urllib
import requests
```

```python
response = requests.get('https://ru.wikipedia.org/w/index.php?title=Всемирная_библиотека_(Норвежский_книжный_клуб)')
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.title)
```

Сначала просматриваем код страницы, и методом пристального взгляда находим, что:
    1. Нужные нам ссылки находятся под тегом <a href ...>
    2. Все внутренние ссылки нужного типа содержат в себе "/wiki/"
    3. Если ссылка на русскую википедию, то она является относительной, а не абсолютной; при открытиим в браузере к ним прибавляется префикс "https://ru.wikipedia.org"
Таким образом, используя синтаксис beautifulSoup, можно получить все внутренние ссылки на странице следующим образом: 

```python
for link in soup.findAll(['a', 'href']):
    if "/wiki/" in str(link):
        print(link['href'])
```

Чтобы не усложнять структуру решения, давайте примем допущение, что нас инересуют только ссылки на русскую википедию/википедию того же языка, что первоначальная ссылка (с иностранными вариантами решается аналогично)

```python
for link in soup.findAll(['a']):
    if 'href' in link.attrs and link['href'][:6] == '/wiki/':
        print(link['href'])
```

Выше вы видите набор непонятных символов, это является кодировкой русского языка. Стоит вам вставить такую ссылку в браузер (приписав  "https://ru.wikipedia.org"), буквы преобразуются в русские.

```python
#запишем ссылки в массив
internal_links = []
for link in soup.findAll(['a']):
    if 'href' in link.attrs and link['href'][:6] == '/wiki/':
        internal_links.append(link['href'])
```

```python
#Выбрать i-ую ссылку, соответственно:
i = 3
print(internal_links[i])

#создать ссылку для перехода можно простой склейкой c префиксом
working_link = "https://ru.wikipedia.org" + internal_links[i]
print(working_link)
```

```python
#посвящается любителям функционального программирования, лаконичный способ отфильтровать ссылки и привести их в порядок
all_links = soup.findAll(['a'])
inner_links = list(filter(lambda x: 'href' in x.attrs and x['href'][:6] == '/wiki/', all_links))
inner_links_clean_ready = list(map(lambda x: 'https://ru.wikipedia.org' + x['href'], inner_links))
```

```python
for link in inner_links_clean_ready:
    print(link)
```

Теперь все готово, чтобы из кусочков собрать наши задачи:
    - знаем, как получать все внутренние ссылки
    - знаем, как по этой ссылке открывать код страницы
    - повторяем это сначала нужное число раз

```python
#сначала напишем функцию, которая по данной странице получает i-ую ссылку на другую страницу википедии; 
#будем использовать ее и с первой и во второй задаче

def get_inner_links(wiki_link):
    response = requests.get(wiki_link)
    soup = BeautifulSoup(response.text, 'html.parser')
    internal_links = []
    for link in soup.findAll(['a']):
        if 'href' in link.attrs and link['href'][:6] == '/wiki/' and 'jpg' not in link['href']:
            working_link = "https://ru.wikipedia.org" + link['href']
            internal_links.append(working_link)
    return (internal_links)
```

```python
#тест
inner_links = get_inner_links('https://ru.wikipedia.org/wiki/Питоны')
for link in inner_links:
    print(link)
```
